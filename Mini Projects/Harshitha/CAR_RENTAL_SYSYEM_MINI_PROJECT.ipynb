##CAR RENTAL SYSYEM


```python

class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password
        self.reservations = []
#here user can do process on reservation like adding reservation and view or cancel your reservation car rental
    def add_reservation(self, car, start_date, end_date):
        self.reservations.append({"car": car, "start_date": start_date, "end_date": end_date})

    def view_reservations(self):
        if not self.reservations:
            print("You have no reservations.")
            return
        print("Your reservations:")
        for i, reservation in enumerate(self.reservations, 1):
            print(f"{i}. Car: {reservation['car'].brand} {reservation['car'].model}, Start Date: {reservation['start_date']}, End Date: {reservation['end_date']}")

    def cancel_reservation(self, index):
        if index < 1 or index > len(self.reservations):
            print("Invalid reservation index.")
            return
        print(f"Cancelling reservation {index}: Car {self.reservations[index - 1]['car'].brand} {self.reservations[index - 1]['car'].model}")
        del self.reservations[index - 1]
#this class shows information about cars like it price or model
class Car:
    def __init__(self, brand, model, car_type, price_per_day, availability=True):
        self.brand = brand
        self.model = model
        self.car_type = car_type
        self.price_per_day = price_per_day
        self.availability = availability
        self.reviews = []
        self.average_rating = 0

    def get_details(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Type: {self.car_type}")
        print(f"Price per day: ${self.price_per_day}")
        print(f"Availability: {'Available' if self.availability else 'Not available'}")

        if self.reviews:
            print("Reviews:")
            for review in self.reviews:
                print(f"Rating: {review['rating']}, Comment: {review['comment']}")
        else:
            print("No reviews yet.")

    def add_review(self, rating, comment):
        self.reviews.append({"rating": rating, "comment": comment})
        self.calculate_average_rating()

    def calculate_average_rating(self):
        if self.reviews:
            total_rating = sum(review['rating'] for review in self.reviews)
            self.average_rating = total_rating / len(self.reviews)

class CarRentalSystem:
    def __init__(self):
        self.users = {}
        self.cars = []

    def register_user(self, username, password):#user registertion process
        if username in self.users:
            print("Username already exists. Please choose another username.")
            return False
        self.users[username] = User(username, password)
        print("Registration successful. You can now login.")
        return True

    def login(self, username, password):#login to entrr into application
        if username not in self.users or self.users[username].password != password:
            print("Invalid username or password. Please try again.")
            return False
        print("Login successful. Welcome back,", username)
        return True

    def add_car(self, brand, model, car_type, price_per_day, availability=True):
        car = Car(brand, model, car_type, price_per_day, availability)
        self.cars.append(car)
        print("Car added successfully.")

    def browse_cars(self, filter=None, sort_by=None):#browe your car from given list and filter them and sort
        filtered_cars = self.cars
        if filter:
            filtered_cars = [car for car in self.cars if car.car_type == filter or car.brand == filter]
        if sort_by:
            if sort_by == 'price':
                filtered_cars.sort(key=lambda car: car.price_per_day)
            elif sort_by == 'brand':
                filtered_cars.sort(key=lambda car: car.brand)
        for car in filtered_cars:
            car.get_details()
    def calculate_rental_fee(self, car_type, duration):#caluclate the total rental fee
        # Simulated fee calculation based on car type and duration
        if car_type == "Sedan":
            return duration * 50  # $50 per day for Sedan
        elif car_type == "SUV":
            return duration * 70  # $70 per day for SUV
        elif car_type == "Truck":
            return duration * 80  # $80 per day for Truck
        elif car_type == "Car":
            return duration * 110  # $110 per day for Car
        else:
            return None

    def make_payment(self, username, rental_fee):

        print(f"Payment of ${rental_fee} received from user {username}. Thank you for using our service!")

    def notify_user(self, username, message):

        print(f"Notification sent to user {username}: {message}")

    def add_review_for_car(self, username, brand, model, rating, comment):#add review of customer who ever bys the car
        car = next((car for car in self.cars if car.brand == brand and car.model == model), None)
        if car:
            car.add_review(rating, comment)
            print(f"Review added for {brand} {model} by {username}.")
        else:
            print(f"Car {brand} {model} not found.")

#Customer Support: The system should provide a way for users to contact customer support in case they face any issues or have any queries.

    def contact_customer_support(self, username, issue):
        print(f"Customer support contacted by user {username} regarding the following issue:\n{issue}")

# Main function
if __name__ == "__main__":
    car_rental_system = CarRentalSystem()

    #we can add cars
    car_rental_system.add_car("Toyota", "Corolla", "Sedan", 50)
    car_rental_system.add_car("Honda", "Civic", "Sedan", 60)
    car_rental_system.add_car("Toyota", "RAV4", "SUV", 70)
    car_rental_system.add_car("Ford", "F-150", "Truck", 80)
    car_rental_system.add_car("BMW", "X-5", "Car", 110)
    while True:
        print("\n1. Register\n2. Login\n3. Browse Cars\n4. Calculate Rental Fee\n5. Add Review for Car\n6. Contact Customer Support\n7. Exit")
        choice = input("Enter your choice: ")
#giving choice u can choose steps entering into application
        if choice == '1':
            username = input("Enter username: ")
            password = input("Enter password: ")
            car_rental_system.register_user(username, password)
        elif choice == '2':
            username = input("Enter username: ")
            password = input("Enter password: ")
            car_rental_system.login(username, password)
        elif choice == '3':
            print("\nAll Cars:")
            car_rental_system.browse_cars()

            print("\nSedans:")
            car_rental_system.browse_cars(filter="Sedan")

            print("\nSorted by price:")
            car_rental_system.browse_cars(sort_by="price")
            print("\nSorted by brand:")
            car_rental_system.browse_cars(sort_by="brand")
        elif choice == '4':
            car_type = input("Enter car type (Sedan, SUV, Truck,Car): ")
            duration = int(input("Enter rental duration (in days): "))
            rental_fee = car_rental_system.calculate_rental_fee(car_type, duration)
            if rental_fee is not None:
                print(f"Estimated rental fee for {car_type} for {duration} days: ${rental_fee}")
            else:
                print("Invalid car type.")
        elif choice == '5':
            username = input("Enter your username: ")
            brand = input("Enter the car brand: ")
            model = input("Enter the car model: ")
            rating = float(input("Enter your rating (out of 5): "))
            comment = input("Enter your review comment: ")
            car_rental_system.add_review_for_car(username, brand, model, rating, comment)
        elif choice == '6':
            username = input("Enter your username: ")
            issue = input("Enter your issue or query: ")
            car_rental_system.contact_customer_support(username, issue)
        elif choice == '7':
            print("Exiting the car rental system. Thank you!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

```

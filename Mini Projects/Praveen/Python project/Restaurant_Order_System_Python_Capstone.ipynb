```python

class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Order:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def calculate_total(self):
        total = sum(item.price for item in self.items)
        return total

class Table:
    def __init__(self, number):
        self.number = number
        self.order = None

def display_menu(menu):
    print("Menu:")
    for i, item in enumerate(menu, 1):
        print(f"{i}. {item.name}: Rs.{item.price}")

def take_order(menu):
    order = Order()
    while True:
        display_menu(menu)
        choice = input("Enter item number to order (0 to finish): ")
        if choice == '0':
            break
        try:
            choice = int(choice)
            if choice < 0 or choice > len(menu):
                raise ValueError
            item = menu[choice - 1]
            order.add_item(item)
            print(f"{item.name} added to the order.")
        except ValueError:
            print("Invalid input. Please enter a valid item number.")
    return order

def add_menu_item(menu):
    name = input("Enter the name of the new menu item: ")
    price = float(input("Enter the price of the new menu item: "))
    menu.append(MenuItem(name, price))
    print(f"{name} added to the menu.")

def main():
    menu = [MenuItem("Roti", 20.00),
MenuItem("Maggi", 20.00),
MenuItem("Upma", 30.00),
MenuItem("Idli", 30.00),
MenuItem("Vada", 35.00),
MenuItem("Poori", 40.00),
MenuItem("Plane Dosa", 35.00),
MenuItem("Onion Dosa", 40.00),
MenuItem("Egg Dosa", 45.00),
MenuItem("Bread Omelette", 40.00),
MenuItem("French Fries", 80.00),
MenuItem("Burger", 120.00),
MenuItem("Pizza", 150.00),MenuItem("Salad", 100.00),
MenuItem("Dal Curry with Rice", 120.00),
MenuItem("Egg Curry with Rice", 140.00),
MenuItem("Veg Noodles", 70.00),
MenuItem("Egg Noodles", 90.00),
MenuItem("Veg Fried Rice", 70.00),
MenuItem("Egg Fried Rice", 70.00),
MenuItem("Veg Manchurian", 70.00),
MenuItem("Chicken Manchurian", 100.00)]
    tables = [Table(i) for i in range(1,11)]
    print("Welcome to our Restaurant!")
    while True:
        print("\n1. Place an Order")
        print("2. Add a new Menu Item")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            while True:
                print("\nAvailable Tables:")
                for table in tables:
                    if table.order is None:
                        print(f"Table {table.number} - Available")
                    else:
                        print(f"Table {table.number} - Occupied")

                table_num = input("\nEnter table number to place an order (0 to exit): ")
                if table_num == '0':
                    break

                try:
                    table_num = int(table_num)
                    if table_num < 1 or table_num > len(tables):
                        raise ValueError
                    table = tables[table_num - 1]
                    if table.order is not None:
                        print("Table already occupied. Please select another table.")
                        continue
                    print(f"\nTable {table_num} selected.")
                    order = take_order(menu)
                    total = order.calculate_total()
                    print(f"Total amount to pay: Rs.{total:.2f}")

                    # Writing order details to a file
                    with open("orders.txt", "w") as file:
                        file.write("Order Details:\n")
                        for item in order.items:
                            file.write(f"{item.name}: Rs.{item.price}\n")
                        file.write(f"Total: Rs.{total:.2f}")
                    print("Order details written to file 'orders.txt'.")

                    table.order = order
                    print("Order placed successfully.")
                except ValueError:
                    print("Invalid input. Please enter a valid table number.")

        elif choice == '2':
            add_menu_item(menu)

        elif choice == '3':
            break

    print("\nExiting Restaurant Order System.")

if __name__ == "__main__":
    main()
```
